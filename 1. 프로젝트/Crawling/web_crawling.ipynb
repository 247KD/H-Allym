```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import openpyxl
import requests

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions['B'].width = 100
# 엑셀 파일 만들기

num = 0
excel_sheet.append(['번호', '제목', '작성자','첨부파일', '등록일'])
#엑셀 첫번째 열에 탭별 이름 부여

url = 'https://www.hallym.ac.kr/hallym_univ/sub05/cP3/sCP1.html'
#전체공지사항 주소
html = requests.get(url).text
soup = BeautifulSoup(html, 'html.parser')
#beautifulsoup로 파싱
data = soup.find_all('li', 'tbl-row')

for item in data:
    item_num = item.find('span','col col-1 tc').text    #번호
    item_title = item.find('a').text                    #제목
    item_name = item.find('span', 'col col-3 tc').text  #작성자
    item_file = item.find('span', 'col col-4 tc').text  #첨부파일
    item_date = item.find('span', 'col col-5 tc').text  #등록일
    excel_sheet.append([item_num, item_title, item_name, item_file, item_date])

cell_A1 = excel_sheet['A1']
cell_A1.alignment = openpyxl.styles.Alignment(horizontal="center")
#시트에 중앙정렬해서 요소 집어넣음

cell_B1 = excel_sheet['B1']
cell_B1.alignment = openpyxl.styles.Alignment(horizontal="center")

cell_C1 = excel_sheet['C1']
cell_C1.alignment = openpyxl.styles.Alignment(horizontal="center")

cell_D1 = excel_sheet['D1']
cell_D1.alignment = openpyxl.styles.Alignment(horizontal="center")

cell_E1 = excel_sheet['E1']
cell_E1.alignment = openpyxl.styles.Alignment(horizontal="center")

excel_file.save('hanallym.xlsx')
excel_file.close()
```
